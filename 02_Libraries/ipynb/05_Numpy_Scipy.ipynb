```python
import numpy as np
import scipy as scipy
import matplotlib.pyplot as plt
```

#### In numpy we create matricies by using lists. Each list contains a row in the matrix

```python
A = np.array([[42, 30], [1, 1]])
B = np.array([[18600],[500]])
```

#### Numpy has a linear algebra solver called linalg.solve we can use to solve for X. 

#### Notice that we don't need to compute the inverse of A. It's needed to solve the problem, but numpy will do it for us, behind the scences, in the solve method

```python
np.linalg.solve(A, B)
```

#### Numpy says the values of x and y that solve these equations are x=300 and y=200
#### Let's just verify that for ourselves. 
#### Recall that our first equation was x + y = 500. Does that work?

```python
300 + 200
```

#### The second equation was 42x + 30y = 18600. Does that work?

```python
42*300 + 30*200
```

#### Numpy hid the steps from us and just output the result. Most of the time this is all we want. However, if we do need to find to get the inverse of a matrix or the determinant there are methods to do so

```python
np.linalg.inv(A) # inverse of A
```

```python
np.linalg.det(A) # determinant of A, an intermediate step needed to get inverse A
```

#### Now we have 3 variables (crops).

#### x + y + z = 200 (land usage)
#### 40x + 60y + 80z = 12600 (money usage)
#### 20x + 25y + 40z = 5950 (labor usage)

```python
A = np.array([[1,1,1],[40,60,80],[20,25,40]])
B = np.array([[200],[12600],[5950]])
```

```python
np.linalg.solve(A, B)
```
