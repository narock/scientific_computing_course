```python
import numpy as np
import matplotlib.pyplot as plt

y = np.loadtxt("piano.txt",float)
x = np.arange(0, len(y), 1)
plt.plot(x,y)
plt.ylabel('Amplitude')
plt.xlabel('Time')
```

```python
import scipy.fftpack
fft = scipy.fftpack.fft(y)
freqs = scipy.fftpack.fftfreq(len(fft)) * 44000
```

```python
plt.plot( freqs, abs(fft) )
plt.xlim(0,3000)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
```

```python
plt.plot( freqs, abs(fft) )
plt.xlim(500,540)
```

```python
import scipy.io.wavfile

rate, audio = scipy.io.wavfile.read('nightingale.wav')
audio = np.mean(audio, axis=1)

N = audio.shape[0]
L = N / rate

print(f'Audio length: {L:.2f} seconds')

plt.plot(np.arange(N) / rate, audio)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
```

```python
a=audio[40000:50000]
N = a.shape[0]
L = N / rate

print(f'Audio length: {L:.2f} seconds')
plt.plot(np.arange(N) / rate, a)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
```

```python
fft = scipy.fftpack.rfft(a)
freqs = scipy.fftpack.rfftfreq(len(fft)) * rate
```

```python
plt.plot(freqs, np.abs(fft))
plt.xlabel('Frequency in Hertz [Hz]')
plt.ylabel('Magnitude')
plt.xlim(0,5000)
```

```python
import scipy.signal

freqs, times, Sx = scipy.signal.spectrogram(a, fs=rate, window='hanning',
                                            nperseg=1024, noverlap=924,detrend=False,
                                            scaling='spectrum')

plt.pcolormesh(times, freqs/1000, 10 * np.log10(Sx), cmap='viridis')
plt.ylabel('Frequency [kHz]')
plt.xlabel('Time [s]')
```

```python
import pandas as pd
data_url = "http://narock.github.io/teaching/CST-411/sunspots.txt"
data = pd.read_csv('sunspots.txt', sep = '\s+')
data.head()
```

```python
plt.plot(data['index'],data['sunspots'])
plt.xlabel('Month')
plt.ylabel('# Sunspots')
```

```python
fft = scipy.fftpack.rfft(data['sunspots'])
freqs = scipy.fftpack.rfftfreq(len(fft)) 
plt.plot( 1./freqs, abs(fft) )
plt.xlim(0,500)
```
