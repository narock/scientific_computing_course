```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1950,2025,1)
y1 = 0.15*x - 299.55 
y2 = 0.02*x - 38.5
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlim(2005,2010)
plt.ylim(0,3)
plt.grid()
plt.legend(['China','US'])
```

```python
a=np.array([[0.15,-1],[0.02,-1]])
b=np.array([299.55,38.5])
print(np.linalg.solve(a,b))
```
